```python
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables
```

```python
rsq_r=smf.ols("RDS~ADMN+MKTS",data=stups1).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADMN~RDS+MKTS",data=stups1).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MKTS~RDS+ADMN",data=stups1).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
data={'Variables':['RDS','ADMN','MKTS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(data)
Vif_df
```

```python
## vif value is less than 20, so no collinearity all variables consider in regression
```
